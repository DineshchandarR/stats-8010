## Linear Regression
se = B0; sxx = B0,0

```python
import numpy as np
import statistics as stat
import math
```

```python
a = '20 20 20 43 68 7.5 7 8 9 10'
b = '1.10 1.06 1.04 4.94 10.37 -0.50 -1.38 -0.72 -0.99 -0.68'
```

```python
def to_array(a):
    val = []
    for i in a.split(' '):
        val.append(float(i))
    return val
```

```python
a = np.array(to_array(a))
n1 = len(a)
b = np.array(to_array(b))
n2 = len(b)
print (f'N1:{n1} and N2:{n2}')
```

```python
sxx = sum((a - stat.mean(a)) ** 2)
sxy = sum(a * b) - (sum(a) * sum(b) / n1)
syy = sum(b ** 2) - (sum(b) ** 2 / n2)
print(f"sxx:{sxx}, sxy:{sxy}, syy:{syy}")
```

```python
r = sxy / math.sqrt(sxx * syy)
r2 = r ** 2
r2
```

```python
B1 = sxy / sxx
B1
```

```python
B0 = stat.mean(b) - (stat.mean(a) * B1)
B0
```

```python
-2.478 + (0.1836*300)
```

### wt(Ht) = B0 + B1(Ht)
Above forumlae will allow us to predict X for Y values based on slope and intercept (LR pred)

```python
s2e = ((1 - r2) * syy) / (n1 - 2)
s2e
```

```python
math.sqrt(s2e)
```

### To test hypotheses about b₀

Upper – tailed       | Lower – tailed        | Two – tailed
-----------------    |--------------------   | ------------------
Ho  :  b0 = b0,0     | Ho  :  b0 = b0,0      | Ho  : b0  = b0,0   
Ha  :  b0 > b0,0     | Ha  :  b0 < b0,0      | Ha  : b0 ≠ b0,0      

assume se = b0 and b0,0 = sxx


Test statistic :

𝑡 = 𝛽̂0−𝛽0,0/ (𝑠𝜀* math.sqrt(Σ𝑥𝑖**2/(𝑛𝑆xx)) (under Ho, is t with n-2 df)


```python
t1 = math.sqrt(s2e) / math.sqrt(sxx)
tobs = B1 / t1
tobs
```

```python
alpha = 0.05
df = n1 - 2
```

```python
from scipy.stats import t
t.ppf(1 - (alpha/2), df)
```

### Confidence Intervals for b0
A 100x(1-alpha)% CI for b0 is

𝛽̂0±(𝑡𝛼/2,𝑛−2)*𝑠𝛽̂0 , where 𝑠𝛽̂0=𝑠𝜀(math.sqrt(Σ𝑥𝑖^2/𝑛𝑆xx))

```python
sB0 = math.sqrt(s2e) * (math.sqrt(sum(a**2)/sxx))

LCL_B0 = B0 - (t.ppf(1 - (alpha/2), ((n1+n2)-2)) * sB0)
HCL_B0 = B0 + (t.ppf(1 - (alpha/2), ((n1+n2)-2)) * sB0)

print(f'sB0: {sB0} LCL: {LCL_B0} and HCL: {HCL_B0}')
```

### Confidence Intervals for E[Y | X = xo]

𝑦̂0±(𝑡𝛼/2,𝑛−2)𝑠𝜀*(math.sqrt(1/𝑛+((𝑥0−𝑥̅)**2/sxx)) , where 𝑦̂0=𝛽̂0+(𝛽̂1*𝑥0)

```python
x0 = 55 # will be provided in the question, x0 = CI for mean at a value (x0)
y0 = B0 + (B1*x0)
T = t.ppf(1 - (alpha/2), (n1-2))
LCL = y0 - (T* math.sqrt(s2e) *(math.sqrt( (1 / n1) + ((x0 - stat.mean(a) )**2 / sxx))))
HCL = y0 + (T* math.sqrt(s2e) *(math.sqrt( (1 / n1) + ((x0 - stat.mean(a) )**2 / sxx))))

print(f'T: {T}, x0: {x0} ,  y0:{y0}, LCL:{LCL}, HCL:{HCL}')
```
